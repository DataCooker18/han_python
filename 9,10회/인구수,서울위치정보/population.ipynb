```python
import pandas as pd
import json
import folium

# 구 별 인구수 데이터 로드
popDF = pd.read_csv('pop_kor.csv', encoding='UTF-8')#, index_col='구별')

# 위치 데이터 json 파일 로드
geo_path = './skorea_municipalities_geo.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
popDF
```

```python
geo_str
```

```python
m =folium.Map([37.57672487388627, 127.11519584981606], zoom_start=10)

m.choropleth(geo_data=geo_str, data=popDF, name='Seoul',
            columns =['구별','인구수'], key_on='feature.id',
            fill_color = 'BuPu', fill_opacity=0.7, line_opacity=0.2,
            legend_name = 'Population')
folium.LayerControl().add_to(m)

"""
geo_data : json 파일
data : data 파일
name : 그래프 이름
columns : data의 축 사용
key_on : json 파일의 변수명 지정
fill_color ; 그래프 색
fill_opacity : 채우는 색 투명도
line_opacity : 그래프 구분하는 선 투명도
legend_name : lenged 이름

folium.LayerControl().add_to() 우측 동적 기능 추가
"""
```
